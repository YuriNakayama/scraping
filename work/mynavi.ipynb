```python
import json
import os
from collections import OrderedDict
import shutil

import pandas
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.expected_conditions import presence_of_element_located
from selenium.webdriver.support.ui import WebDriverWait
```

```python
data_path = "/home/jovyan/work/data/"
```

```python
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Remote(command_executor=os.environ["SELENIUM_URL"], options=options)
```

```python
start_url = "https://tenshoku.mynavi.jp/list/"
```

```python
# ドライバとタイムアウト値を指定
driver.get(start_url)
wait = WebDriverWait(driver, 10)
```

```python
elements = driver.find_elements(By.XPATH, '//div[@class="cassetteRecruit__content"]')
```

```python
content_dicts = OrderedDict()
for element in elements:
    content_dict = OrderedDict()
    name = element.find_element(By.XPATH, ".//h3[@class='cassetteRecruit__name']").text.split('|')[0]
    content_dict["href"] = element.find_element(By.XPATH, f'.//a[@class="linkArrowS"]').get_attribute('href')
    table_element = element.find_element(By.XPATH, './/table[@class="tableCondition"]/tbody')
    for table_item in table_element.find_elements(By.XPATH, './tr'):
        content_dict[table_item.find_element(By.XPATH, './th').text] = table_item.find_element(By.XPATH, './td').text
    content_dicts[name] = content_dict
```

```python
output_json_path = data_path + "mynavi/temp/"
os.makedirs(output_json_path, exist_ok=True)
with open(output_json_path+ "temp.json", "w") as f:
    content_json = json.dump(content_dicts, f)
```
